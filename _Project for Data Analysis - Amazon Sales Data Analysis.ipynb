# EDA

## Size

```python
ax=sns.countplot(x='Size', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

##### Note: From the above graph it can be concluded that most people are of size Medium

```python
## Size and Quantity
```

```python
Size_Quan = df.groupby(df['Size'], as_index = False)['Quantity'].sum().sort_values(by = 'Quantity', ascending = False)
```

```python
bx = sns.barplot (x='Size',y='Quantity', data = Size_Quan)
for bars in bx.containers:
    bx.bar_label(bars)
```

##### Note: From the above graph, it is clear that the highest quantity of sales is of size M, closely followed by L and XL.

```python
## Size and Amount
```

```python
Size_Amt = df.groupby(df['Size'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
Size_Amt
```

```python
cx = sns.barplot (x='Size',y='Amount', data = Size_Amt)
sns.set(rc={'figure.figsize':(30,15)})
for bars in cx.containers:
    cx.bar_label(bars)
```

##### Note: From the above graph, it is clear that the higesht amount of sales is of size M closely followed by L and XL.

### Courier Status

```python
dx = sns.countplot(x='Courier Status', data = df, hue = 'Status')
sns.set(rc={'figure.figsize':(12,5)})
for bars in dx.containers:
    dx.bar_label(bars)
```

##### Note: From above graph, we can see that the highest number or orders are shipped through Courier

```python
# histogram
dx = df['Size'].hist()
for bars in dx.containers:
    dx.bar_label(bars)
```

```python
df['Category'] = df['Category'].astype('str')

plt.xlabel('Category')
plt.title('Count of sales for each Category')
plt.margins(x=0.01, y=0.1)
plt.bar_label(bars, fontsize=15, color='blue')
plt.xticks(rotation = 90)
plt.hist(df['Category'], bins = 50, edgecolor = 'black')
plt.figure(figsize=(15, 10))



plt.show()
```

#### Note: From above graph, we can see that the highest Category sold is T-shirt

```python
df['Size'] = df['Size'].astype('str')

plt.xlabel('Size')
plt.title('Count of sales for each Size')
plt.margins(x=0.01, y=0.1)
plt.bar_label(bars, fontsize=15, color='blue')
plt.xticks(rotation = 90)
plt.hist(df['Size'], bins = 50, edgecolor = 'black')
plt.figure(figsize=(15, 10))


plt.show()
```

##### Note: From the above graph we can see that the highest count of sales is in the size M

```python
### Checking B2B data using Pie Chart
B2B_label = df['B2B'].value_counts()
exp=[0,0.2]
plt.pie(df['B2B'].value_counts(), labels = B2B_label.index , explode = exp, autopct = '%1.1f%%', shadow = False)
plt.title('Bifurcation of B2B data')
plt.show()
```

##### Note: From the above Pie, we can understand that 99.2% buyers are retailers and the rest 0.8% buy from B2B channel

```python
### Checking Fulfillment data using Pie Chart
Ful_label = df['Fulfilment'].value_counts()

fig, ax = plt.subplots()
ax.pie(Ful_label,labels = Ful_label.index, radius = 0.7,wedgeprops = {'edgecolor':'White'}, autopct = '%1.1f%%', shadow = False)
ax.set(aspect = 'equal')
plt.show()
```

##### Note: From the above Pie, we can see that all the Fulfilment has been done throught Merchant

```python
### Checking fulfilled_by data using Pie Chart
fulfilled_by_label = df['fulfilled-by'].value_counts()
plt.pie(fulfilled_by_label, labels = fulfilled_by_label.index , autopct = '%1.1f%%', shadow = False)
plt.title('Fulfilled By')
plt.show()
```

##### Note: From above Pie, we can clearly see that all the orders were fulfilled through easy shipments

```python
### Use scatter plot
plt.scatter(df['Category'],df['Size'])
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot of Category and Size')
```

##### Note: From the above graph we can find out exactly for each category, the different types of sizes available.

```python
# Plot Count of cities by State
#plt.figure(figsize=(12,6))
sns.countplot(data =df, x='ship-state')
plt.xticks(rotation = 90)
plt.title('Distribution of Orders in each State')
plt.show()
```

```python
# Top 10 states
top_10_States = df['ship-state'].value_counts().head(10)
top_10_States
```

```python
# Plot count of cities by State
sns.countplot(data =df[df['ship-state'].isin(top_10_States.index)], x='ship-state')
plt.xlabel('Ship State')
plt.ylabel('Count')
plt.xticks(rotation = 60)
plt.title('Distribution of Buyers in each State')
plt.show()
```

##### Note: From above graph, it is clear that people in Maharashtra have a high ourchasing power and most of the buyers are from Maharashtra followed by Karnataka and Telengana.
