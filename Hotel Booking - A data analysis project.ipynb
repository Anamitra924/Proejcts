```python
# Which hotel's cancellations rate is high
plt.figure(figsize=(10,5))
ax1 = sns.countplot(x='hotel', data=df, hue = 'is_canceled', palette = 'Blues')
for bars in ax1.containers:
    ax1.bar_label(bars)
plt.title('Reservation status in different hotels', size = 15)
plt.xlabel('Hotel')
plt.ylabel('Number of reservations')
plt.legend(['Not Canceled','Canceled'])
plt.show()
```

## From the above graph, it is clear that City hotel has faced 3x the cancelations that Resort hotel.

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True) # In Resort hotels only 28% reservations are getting cancelled
```

```python
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True) # In City hotel almsot 42% reservations are getting cancelled.
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# Visualization for avg daily rate for each hotel
plt.figure(figsize=(20,8))
plt.plot(resort_hotel.index,resort_hotel['adr'],label = 'Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label = 'City Hotel')
plt.title('Average daily rate in each hotel', fontsize = 20)
plt.legend(fontsize=20)
plt.show()
```

### From the above graph, we can see that the avg daily rate of resort hotel is much higher than that of city hotel.

```python
# For which month reservation is highest and for which cancellation is highest
df['Month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1 = sns.countplot(x='Month', hue = 'is_canceled', data=df, palette = 'bright')
for bars in ax1.containers:
    ax1.bar_label(bars)
plt.title('Reservation status per month', size = 25)
plt.xlabel('Month')
plt.ylabel('Number of reservations')
plt.legend(['Not cancelled','Cancelled'])
plt.show()
```

### From the above graph, we can conclude:
### highest number of cancellations in January followed by October
### highest number of reservations (not cancelled) was in August followed by July

```python
data = df[df['is_canceled'] == 1].groupby('Month')[['adr']].sum().reset_index()
plt.figure(figsize=(15, 8))
plt.title('Average Daily Rate for each month', fontsize=30)
ax1 = sns.barplot(x='Month', y='adr', data=data)
for bars in ax1.containers:
    ax1.bar_label(bars)
plt.ylabel('Average Daily Rate')
plt.show()
```

### From the above graph that in august ADR is lowest and January ADR is highest. So it proves our hypothesis that, in January higher cancellations occur due to higher ADR

```python
# top 10 countries with reservations cancelled
cancelled_data = df[df['is_canceled'] ==1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(7,6))
plt.title('Top 10 countries with highest number of cancellations')
plt.pie(top_10_country,autopct = '%.2f', labels=top_10_country.index)
plt.show()
```

### Highest percentage of cancellations is in the hotels of Portugal. So clearly, the problem lies there. Hotels is portugal need to improve their services as well as their facilities, improve prices, give promotional discounts, etc.

```python
# We will check whether the ADR of successful reservations are high or cancellations are hogher?
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

not_cancelled_data = df[df['is_canceled'] ==0]

not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize=(25,8))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label = 'Not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label = 'Cancelled')
plt.legend()
```

```python
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016')&(cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016')&(not_cancelled_df_adr['reservation_status_date']<'2017-09')] 
```

```python
plt.figure(figsize=(16,5))
plt.title('Average Daily Rate', fontsize=25)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label = 'Not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label = 'Cancelled')
plt.legend()
```

### From the above graph, it is clear that a higher ADR is causing higher number of cancellations. So, ADR is an influencial factor when it comes to cancellations.
